#### Importing Dependencies:

```python
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
```

#### Open CSV/Dataset:

```python
df = pd.read_csv("poems_for_printing.csv")  
```

#### Apply themes with zero-shot classification:

```python
theme_labels = [
    "Love and Relationships",
    "Nature and the Environment",
    "Loss and Grief",
    "Resilience and Strength",
    "Hope and Optimism",
    "Joy and Celebration",
    "Beauty and Aesthetics",
    "Time and Transience",
    "Spirituality and Wonder"
]
```

```python
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

def classify_poem(text, threshold=0.2):
    result = classifier(text, theme_labels, multi_label=True)
    selected = [label for label, score in zip(result['labels'], result['scores']) if score >= threshold]
    return selected[:2]  
'''Used to classify a poem and pick the top two labels over the threshold'''

themes_list = []
for poem in tqdm(df['text'], desc="Classifying poems"):
    try:
        themes = classify_poem(poem)
    except Exception as e:
        print("Error processing poem:", e)
        themes = []
    themes_list.append(themes)
# Applies this to all of the poems

df['themes'] = themes_list

df.to_csv("poems_with_themes.csv", index=False)
```
