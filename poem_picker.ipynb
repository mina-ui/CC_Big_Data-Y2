```python
import pandas as pd
import random
```

```python
df = pd.read_csv('poems_with_themes.csv')
```

```python
theme_choices = [
    "Love and Relationships",
    "Nature and the Environment",
    "Loss and Grief",
    "Resilience and Strength",
    "Hope and Optimism",
    "Joy and Celebration",
    "Beauty and Aesthetics",
    "Time and Transience",
    "Spirituality and Wonder"
]
```

```python
print("The available themes:")
print(theme_choices)
```

```python
input_theme1 = input("Enter the first theme: ")
input_theme2 = input("Enter the second themes theme: ")
# https://www.w3schools.com/python/ref_func_input.asp
```

```python
filtered_poems = df[
    df['themes'].apply(lambda themes: input_theme1 in themes or input_theme2 in themes)
]
# https://www.geeksforgeeks.org/python/applying-lambda-functions-to-pandas-dataframe/

if filtered_poems.empty:
    print("No poems found with those themes.")
else:
    selected = filtered_poems.sample(n=1).iloc[0]
    # sample(n=1).iloc[0] was with the help of chatgpt (displays a row from the csv at random (but has the two typed themes))
    print(f"Title: {selected['title']}")
    print(f"Author: {selected['author']}")
    print(f"Poem: {selected['text']}")
```
