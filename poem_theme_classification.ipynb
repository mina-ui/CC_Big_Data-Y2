```python
from transformers import pipeline
from pprint import pprint

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
theme_labels = [
  "love and relationships",
  "nature and the environment",
  "loss and grief",
  "resilience and strength",
  "hope and optimisim",
  "joy and celebration",
  "beauty and aesthetics",
  "time and transience",
  "spirituality and wonder"
]

# chatgpt suggested combining like this to help the model understand better
```

```python
THRESHOLD = 0.2  
```

```python
def select_labels(output, threshold=0.2):
    selected = []
    for label, score in zip(output['labels'], output['scores']):
        if score >= threshold:
            selected.append((label, score))
    return selected
```

```python
poem_text = "Nights have a habit of mysterious gifts and refusals, of things half given away, half withheld."

result = classifier(poem_text, theme_labels, multi_label=True)

pprint(result)
```

- Wayyyy more confident than training distilBERT- probably because I didn't have a large enough pre-tagged dataset. 

- Adjusting the themeing labels to have two themes per label increased confidence a lot!
