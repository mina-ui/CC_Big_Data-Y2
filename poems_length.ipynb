#### Filtering Poems by length:
##### (max char 2000, for better printing)

```python
import csv

def is_short_enough(poem_text, max_lines=60, max_chars=2000):
    lines = poem_text.strip().split('\n')
    if len(lines) > max_lines:
        return False
    if len(poem_text) > max_chars:
        return False
    return True

with open('poems_cleaned.csv', 'r', encoding='utf-8') as infile, \
     open('poems_for_printing.csv', 'w', newline='', encoding='utf-8') as outfile:

    reader = csv.DictReader(infile)
    fieldnames = ['title', 'author', 'date', 'text']
    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()

    for row in reader:
        if is_short_enough(row['text']):
            writer.writerow(row)

print("Saved to poems_for_printing.csv")
```
